**Задание 1.**

С помощью функции `plot()` постройте графики функции $y=sin(x)$ ($x=\overline{0, 2π}$) при всех возможных значениях параметра `type`. Отобразите все графики в одном графическом окне.

```r
x <- seq(0, 2 * pi, length.out = 100)
types <- c("p", "l", "b", "o")
par(mfrow = c(2, 2))
lapply(types, function(type) {
  plot(x, sin(x), type = type, main = paste("type =", type))
})
```

**Задание 2.**

С помощью функции `plot()` постройте графики функции $y=sin(x)$ ($x=\overline{0, 2π}$) при всех возможных значениях параметра `lty`. Отобразите все графики в одном графическом окне.


```r
x <- seq(0, 2 * pi, length.out = 100)
ltys <- c(1, 2, 3, 4)
par(mfrow = c(2, 2))
lapply(ltys, function(lty) {
  plot(x, sin(x), lty = lty, main = paste("lty =", lty))
})
```

**Задание 3.**

Выполните задание 2 с использованием функции `split.screen()`. Задайте для каждого из подокон разный цвет фона.


```r
x <- seq(0, 2 * pi, length.out = 100)
ltys <- c(1, 2, 3, 4)
bg_colors <- c("white", "lightgray", "darkgray", "blue")

split.screen(figs = c(2, 2))

for (i in 1:4) {
  screen(i)
  par(bg = bg_colors[i])
  plot(x, sin(x), lty = ltys[i], main = paste("lty =", ltys[i]))
}

close.screen(all = TRUE)
```

**Задание 4.**

Постройте график функции $y(x) = \pi x^2 \log_{10}(x)$, проименуйте оси соответственно. (см. `help(plotmath)` и `demo(plotmath)`). Пусть цвет рамки будет зеленым, а цвет самого графика в RGB палитре соответствует красному. Задайте расстояние между надписями и осями так, чтобы надписи полностью умещались в графическом окне (см. параметр `mgp` функции `plot()`).

```r
x <- seq(0.01, 3, length.out = 100)
y <- pi * x^2 * log10(x)
plot(x, y, type = "l", col = "red", xlab = expression(paste("X")), ylab = expression(paste("Y(X)")), mgp = c(2, 0.5, 0))
box(col = "green")
```

**Задание 5.**

Задайте вектор $x=(-2, -1, 0, 1, 2)$. В одном графическом окне (в 4 подокнах) по точкам $x$ изобразите графически значения функции $y(x) = x^3-3x$  в виде:
- точек,
- линий,
- линий и точек,
- кривой.

Сохраните полученные изображения в файле figure.jpg.

```r
x <- c(-2, -1, 0, 1, 2)
y <- x^3 - 3*x
par(mfrow = c(2, 2))
plot(x, y, type = "p", main = "Точки")
plot(x, y, type = "l", main = "Линии")
plot(x, y, type = "b", main = "Линии и точки")
plot(x, y, type = "s", main = "Кривая")
dev.copy(jpeg, "figure.jpg")
dev.off()
```

Задайте вектор $x=(3, 3.1, 3.2, ... , 6)$. Постройте графики функций $y_1(x)=\pi x$  и $y_2(x) = \exp(x) \cos(x)$ в указанном диапазоне значений аргумента $x$ следующим образом:

а) постройте оба графика разного цвета на одном рисунке с помощью функции `par()`, добавьте легенду и сетку для каждого графика. В чем недостаток такого построения?

б) в новом графическом окне постройте оба графика разного цвета на одном рисунке с помощью графических функций низкого уровня. Сравните с вариантом из пункта а). Добавьте сетку для каждого графика и расположите непрозрачную легенду в верхнем левом углу рисунка. Используя функцию `locator()`, расположите легенду в любом удобном месте рисунка по клику мыши. Обозначьте точку пересечения графиков кружком и надписью "Точка пересечения" по клику мыши.

в) установите пакет `ggplot2`, изучите документацию функции `ggplot()` (`help(ggplot)`) и с ее помощью постройте графики обеих функций на одном рисунке (на разных слоях). Поменяйте названия осей на "X" и "Y(X)".


```r
x <- seq(3, 6, by = 0.1)
y1 <- pi * x
y2 <- exp(x) * cos(x)

print('а)')
plot(x, y1, type = "l", col = "green", xlab = "X", ylab = "Y(X)")
lines(x, y2, type = "l", col = "blue")
legend("topright", legend = c("πx", "exp(x)*cos(x)"), col = c("green", "blue"), lty = 1)
grid()
# Недостаток - функции y1 и y2 имеют разные масштабы, и поэтому график y2 может быть сложно увидеть, когда он находится на том же графике, что и y1.
```

```r
print('б)')
x <- seq(3, 6, by = 0.1)
y1 <- pi * x
y2 <- exp(x) * cos(x)
par(mfrow = c(1, 1))
plot(x, y1, type = "l", col = "green", xlab = "X", ylab = "Y(X)", ylim = range(c(y1, y2)))
lines(x, y2, type = "l", col = "blue")
grid()
legend("topright", legend = c("πx", "exp(x)*cos(x)"), col = c("green", "blue"), lty = 1)
click <- locator(1)
points(click$x, click$y, pch = 19, col = "black")
text(click$x, click$y, "Точка пересечения")
```

```r
print('в)')
library(ggplot2)
x <- seq(3, 6, 0.1)
y1 <- pi * x
y2 <- exp(x) * cos(x)
df <- data.frame(x = x, y1 = y1, y2 = y2)
ggplot(df, aes(x = x)) +
geom_line(aes(y = y1), color = "blue") +
geom_line(aes(y = y2), color = "green") +
xlab("X") +
ylab("Y(X)")
```
