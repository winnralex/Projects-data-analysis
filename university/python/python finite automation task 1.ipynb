Вариант 2. Нерифмованные стихи. Автомат принимает на вход фразу (например на русском), и выделяет слога и строки. По слогам и строкам различает (заканчивает работу в конечном состоянии, названном соответственно) минимум 2 из следующих форм стихов:

1.       Хокку - 17 слогов (5-7-5) : Три строки, 5 слогов в первой, 7 во второй, 5 в третей.

2.       Сэдока - 38 слогов (5+7+7+5+7+7) : Аналогично для 6 строк

3.       Синквейн -  22 слогов (2+4+6+8+2) : Аналогично для 5 строк

4.       Т.п.

```python
# E {0, 1}  алфавит
# E* {e, 0, 00, 01, ....}  множество комбинаций
# L =< E язык
# M = {Q, q0, E, b, F}   автомат
```

```python
class State: # состояние 
	def __init__(self, state_id, edges, is_finish):
		self.state_id = state_id
		self.edges = edges
		self.is_finish = is_finish
	
	def __str__(self):
		return str(self.state_id) + ' : ' + str(self.edges) + ' = ' + str(self.is_finish)
	
	def next(self, symbol):
		ids = []
		for edge in self.edges:
			if edge.startswith(symbol):
				ids.append(edge.split('>')[1])
		return ids
	
	def next_epsilon(self):
		ids = []
		for edge in self.edges:
			if edge.startswith('#e'):
				ids.append(edge.split('>')[1])
		return ids
```

```python
class DFA: # ДКА - детерминированный конечный автомат
	
    def __init__(self, name, states, alphabet):
        self.name = name
        self.alphabet = alphabet
        self.states = states
        self.current_state = self.states[0]
	
    def reset(self):
        self.current_state = self.states[0]
	
    def __str__(self):
        return '#DFA "' + self.name + '" ' + str(self.current_state)
        
    def check_alph(self, word):
        for letter in word:
            if letter in self.alphabet:
                print(letter, 'In alphabet')
            else:
                print(letter, 'Not in alphabet')
                break
	
    def compute(self, word):
        for symbol in word:
            next_ids = self.current_state.next(symbol) # список следующих состояния
			# для ДКА список может быть длины 1 или 0
            if len(next_ids) == 0:
                return False
            for state in self.states:
                if state.state_id == next_ids[0]:
                    self.current_state = state
        print('Finish in ' + str(self.current_state))
        return self.current_state.is_finish
		
class NFA(DFA): # недетерминированный конечный автомат, наследует ДКА
	
	def __str__(self):
		return '#NFA "' + self.name + '" ' + str(self.current_state)
		
	def compute(self, word):
		if word == '':
			return self.current_state.is_finish
		symbol = word[0]
		remainder = word[1:] # выбрасываем из остатка нулевой символ
		next_ids = self.current_state.next(symbol) # список следующих состояния
		ep_next_ids = self.current_state.next_epsilon() # список следующих состояния по эпсилон-переходам
		# для НКА список может быть произвольной длины (0, 1, .. 100..)
		if len(next_ids) + len(ep_next_ids) == 0:
			return False
		for next_id in next_ids:
			for state in self.states:
				if state.state_id == next_id:
					self.current_state = state
					if self.compute(remainder):
						return True
		for next_id in ep_next_ids:
			for state in self.states:
				if state.state_id == next_id:
					self.current_state = state
					if self.compute(word):
						return True
		return False			
#		print('Finish in ' + str(self.current_state))
#		return self.current_state.is_finish
```
