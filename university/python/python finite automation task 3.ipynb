Вариант 2. Торги на бирже. Каждое действие - покупка одного из возможных пакетов ценных бумаг. У каждого пакета есть своя цена и случайный разброс её изменения в день. Изменение цен происходит после покупки и сразу после этого происходит продажа пакета по новой цене. Цены не сбрасываются на стартовый уровень между торгами. Реализовать вариант без комиссии и с комиссией.

```python
from random import random, randint
# у нас есть какое-то количество одноруких бандитов (разных действий)
bandits = [(0.01, 220), (0.02, 130), (0.015, 100), (0.005, 450)]
means = [] # мат ожидания выигрыша
for i in range(len(bandits)):
	means.append(bandits[i][0]*bandits[i][1])

def playAB(out = False):
	T = 200000 # общее количество игр
	
	# А/В тесты - сыграем N раз на каждом автомате, доиграем на лучшем
	N = 10000
	M = T - N * len(bandits)

	wins = [0] * len(bandits)
	for i in range(len(bandits)):
		for _ in range(N):
			wins[i] += bandits[i][1] if random() < bandits[i][0] else 0
	
	max_id = wins.index(max(wins))
	for _ in range(M):
		wins[max_id] += bandits[max_id][1] if random() < bandits[max_id][0] else 0

	total = sum(wins)
	if out:
		print('А/В тесты:')
		print('A/B tests:' , wins)
		print('Total:', total)
		print('Maximum:', T * max(means)) 
		print('Упущеная выгода:', (1 - total / (T * max(means)))*100, '%')
	return (1 - total / (T * max(means)))*100

def playEGreedy(out = False):
	T = 200000 # общее количество игр
	# е-жадный алгоритм : с вероятностью e-используем, и обратной исследуем
	
	plays = [0] * len(bandits) # сколько играли
	wins = [0] * len(bandits) # сколько выиграли
	rates = [0] * len(bandits) # соотношение 

	e = 0.8 # эпсилон

	for _ in range(T):
		if random() < e: # используем (exploit)
			max_id = rates.index(max(rates)) # индекс лучшего
			wins[max_id] += bandits[max_id][1] if random() < bandits[max_id][0] else 0
			plays[max_id] += 1
			rates[max_id] = wins[max_id] / plays[max_id]
		else:
			rand_id = randint(0, len(bandits)-1)		
			wins[rand_id] += bandits[rand_id][1] if random() < bandits[rand_id][0] else 0
			plays[rand_id] += 1
			rates[rand_id] = wins[rand_id] / plays[rand_id]

	total = sum(wins)
	if out:
		print('E-жадный алгоритм:')
		print('Total:', total)
		print('Maximum:', T * max(means)) 
		print('Упущеная выгода:', (1 - total / (T * max(means)))*100, '%')
	return (1 - total / (T * max(means)))*100
	
def playEGUCB(out = False):
	T = 200000 # общее количество игр
	# е-жадный алгоритм : с вероятностью e-используем, и обратной исследуем
	
	plays = [0] * len(bandits) # сколько играли
	wins = [0] * len(bandits) # сколько выиграли
	rates = [0] * len(bandits) # соотношение 
	L = 10 # доверительный интервал - "до скольки поражений играем"
	e = 0.99 # эпсилон

	while T > sum(plays):
		if random() < e: # используем (exploit)
			max_id = rates.index(max(rates)) # индекс лучшего
			#wins[max_id] += bandits[max_id][1] if random() < bandits[max_id][0] else 0
			#plays[max_id] += 1
			loss = 0
			while loss < L and T > sum(plays):
				if random() < bandits[max_id][0]:
					wins[max_id] += bandits[max_id][1]
				else:
					loss += 1 
				plays[max_id] += 1
			rates[max_id] = wins[max_id] / plays[max_id]
		else:
			rand_id = randint(0, len(bandits)-1)
			loss = 0
			while loss < L and T > sum(plays):
				if random() < bandits[rand_id][0]:
					wins[rand_id] += bandits[rand_id][1]
				else:
					loss += 1 
				plays[rand_id] += 1
			rates[rand_id] = wins[rand_id] / plays[rand_id]

	total = sum(wins)
	if out:
		print('E-жадный алгоритм:')
		print('Total:', total)
		print('Maximum:', T * max(means)) 
		print('Упущеная выгода:', (1 - total / (T * max(means)))*100, '%')
	return (1 - total / (T * max(means)))*100
	
resAB = [0]
resEG = []
resEGUCB = []
for _ in range(100):
	#resAB.append(playAB())
	resEG.append(playEGreedy())
	resEGUCB.append(playEGUCB())
	
print('Max AB:', max(resAB), 'Max EG:', max(resEG), 'Max UCB:', max(resEGUCB))
print('Min AB:', min(resAB), 'Min EG:', min(resEG), 'Min UCB:', min(resEGUCB))
print('Avr AB:', sum(resAB)/len(resAB), 'Avr EG:', sum(resEG)/len(resEG), 'Avr UCB:', sum(resEGUCB)/len(resEGUCB))
```

```python
from random import uniform, choice, randint
from statistics import mean

# Наборы пакетов акций с начальными ценами и процентными изменениями

investments = [
    (120, 0.03),  
    (95, 0.01), 
    (105, 0.01), 
    (130, 0.01), 
    (110, 0.01)
]

# investments = [
#     (115, 0.005), 
#     (110, 0.006), 
#     (118, 0.005), 
#     (121, 0.004), 
#     (117, 0.005)
# ]



# investments = [
#     (90, 0.03),
#     (110, 0.01), (120, 0.01), (100, 0.01), (130, 0.01),
#     (115, 0.01), (105, 0.01), (135, 0.01), (125, 0.01),
#     (112, 0.01), (122, 0.01), (132, 0.01), (101, 0.01),
#     (102, 0.01), (110, 0.01), (130, 0.01), (115, 0.01),
#     (105, 0.01), (134, 0.01), (124, 0.01)
# ]

# investments = [
#     (102, 0.005), (113, 0.006), (117, 0.004), (116, 0.005), 
#     (119, 0.005), (111, 0.006), (120, 0.005), (123, 0.005), 
#     (130, 0.004), (108, 0.005), (110, 0.005), (114, 0.006), 
#     (122, 0.005), (118, 0.004), (127, 0.006), (125, 0.005),
#     (121, 0.005), (115, 0.006), (109, 0.004), (105, 0.005)
# ]

# Устанавливаем комиссию как процент от операции
commission_rate = 0.005  

def apply_commission(price):
    # Учитываем комиссию за торговую операцию
    return price * (1 - commission_rate)

def evaluate_AB_strategy():
    cash = 5000  # Начальный капитал
    duration = 3650  # Количество торговых дней
    target = 20000  # Финансовая цель
    segment = 500  # Количество дней для первичной подготовки данных

    results = [0] * len(investments)
    operations = [0] * len(investments)

    # Первоначальная фаза
    for idx in range(len(investments)):
        for _ in range(segment):
            if cash >= investments[idx][0]:
                # Учитываем комиссию при покупке
                cost_with_commission = apply_commission(investments[idx][0])
                cash -= cost_with_commission
                fluctuation = uniform(-investments[idx][1], investments[idx][1])
                updated_price = investments[idx][0] * (1 + fluctuation)
                # Учитываем комиссию при продаже
                revenue_with_commission = apply_commission(updated_price)
                results[idx] += revenue_with_commission - cost_with_commission
                cash += revenue_with_commission
                investments[idx] = (updated_price, investments[idx][1])
                operations[idx] += 1

    top_idx = results.index(max(results))

    # Реверсируем оставшееся количество дней после сегмента
    remaining_days = duration - segment * len(investments)

    # Фаза эксплуатации
    while remaining_days > 0:
        remaining_days -= 1

        if cash >= target or cash <= 0:
            break

        if cash >= investments[top_idx][0]:
            # Учитываем комиссию при покупке
            cost_with_commission = apply_commission(investments[top_idx][0])
            cash -= cost_with_commission
            fluctuation = uniform(-investments[top_idx][1], investments[top_idx][1])
            updated_price = investments[top_idx][0] * (1 + fluctuation)
            # Учитываем комиссию при продаже
            revenue_with_commission = apply_commission(updated_price)
            results[top_idx] += revenue_with_commission - cost_with_commission
            cash += revenue_with_commission
            investments[top_idx] = (updated_price, investments[top_idx][1])
            operations[top_idx] += 1

    return cash, operations[top_idx]

def evaluate_EG_strategy():
    cash = 5000
    duration = 3650
    target = 20000
    e_factor = 0.9

    actions = [0] * len(investments)
    profits = [0] * len(investments)
    averages = [0] * len(investments)

    for _ in range(duration):
        if cash >= target or cash <= 0:
            break

        if uniform(0, 1) < e_factor:
            chosen_idx = averages.index(max(averages))
            if cash >= investments[chosen_idx][0]:
                cash -= investments[chosen_idx][0]
                fluctuation = uniform(-investments[chosen_idx][1], investments[chosen_idx][1])
                updated_price = investments[chosen_idx][0] * (1 + fluctuation)
                transactions = apply_commission(updated_price - investments[chosen_idx][0])
                profits[chosen_idx] += transactions
                cash += apply_commission(updated_price)
                investments[chosen_idx] = (updated_price, investments[chosen_idx][1])
                actions[chosen_idx] += 1
                averages[chosen_idx] = profits[chosen_idx] / actions[chosen_idx]
        else:
            while True:
                random_choice = choice(range(len(investments)))
                if cash >= investments[random_choice][0]:
                    break
            cash -= investments[random_choice][0]
            fluctuation = uniform(-investments[random_choice][1],
investments[random_choice][1])
            updated_price = investments[random_choice][0] * (1 + fluctuation)
            transactions = apply_commission(updated_price - investments[random_choice][0])
            profits[random_choice] += transactions
            cash += apply_commission(updated_price)
            investments[random_choice] = (updated_price, investments[random_choice][1])
            actions[random_choice] += 1
            averages[random_choice] = profits[random_choice] / actions[random_choice]

    top_choice = profits.index(max(profits))
    return cash, actions[top_choice]

def evaluate_UCB_strategy():
    cash = 5000
    duration = 3650
    target = 20000
    e_factor = 0.9
    limit = 3  # Порог доверия (лимит убытков), после которого пакет считается временно небезопасным

    records = [0] * len(investments)
    rewards = [0] * len(investments)
    mean_rewards = [0] * len(investments)

    for _ in range(duration):
        if cash >= target or cash <= 0:
            break

        if uniform(0, 1) < e_factor:
            best_choice = mean_rewards.index(max(mean_rewards))
            decline_count = 0

            while decline_count < limit:
                if cash < investments[best_choice][0]:
                    if all(cash < inv[0] for inv in investments):
                        return cash, sum(records)
                    break
                
                cash -= investments[best_choice][0]
                fluctuation = uniform(-investments[best_choice][1], investments[best_choice][1])
                updated_price = investments[best_choice][0] * (1 + fluctuation)
                cash += apply_commission(updated_price)

                if fluctuation >= 0:
                    rewards[best_choice] += apply_commission(updated_price - investments[best_choice][0])
                else:
                    decline_count += 1
                
                investments[best_choice] = (updated_price, investments[best_choice][1])
                records[best_choice] += 1
                mean_rewards[best_choice] = rewards[best_choice] / records[best_choice]

        else:
            selection_idx = randint(0, len(investments) - 1)
            decline_count = 0

            while decline_count < limit:
                if cash < investments[selection_idx][0]:
                    if all(cash < inv[0] for inv in investments):
                        return cash, sum(records)
                    break

                cash -= investments[selection_idx][0]
                fluctuation = uniform(-investments[selection_idx][1], investments[selection_idx][1])
                updated_price = investments[selection_idx][0] * (1 + fluctuation)
                cash += apply_commission(updated_price)

                if fluctuation >= 0:
                    rewards[selection_idx] += apply_commission(updated_price - investments[selection_idx][0])
                else:
                    decline_count += 1

                investments[selection_idx] = (updated_price, investments[selection_idx][1])
                records[selection_idx] += 1
                mean_rewards[selection_idx] = rewards[selection_idx] / records[selection_idx]

    top_selection = rewards.index(max(rewards))
    return cash, records[top_selection]

# Запуск симуляций для всех стратегий
results_AB = []
results_EG = []
results_UCB = []

steps_AB = []
steps_EG = []
steps_UCB = []

# Проведение симуляции 100 раз
for _ in range(100):
    final_AB, steps_AB_ = evaluate_AB_strategy()
    final_EG, steps_EG_ = evaluate_EG_strategy()
    final_UCB, steps_UCB_ = evaluate_UCB_strategy()

    results_AB.append(final_AB)
    results_EG.append(final_EG)
    results_UCB.append(final_UCB)

    steps_AB.append(steps_AB_)
    steps_EG.append(steps_EG_)
    steps_UCB.append(steps_UCB_)

# Итоги симуляций
print('Стратегия A/B: Максимальный капитал = ${}, Среднее количество операций = {}'.format(max(results_AB), mean(steps_AB)))
print('Стратегия ε-жадная: Максимальный капитал = ${}, Среднее количество операций = {}'.format(max(results_EG), mean(steps_EG)))
print('Стратегия ε-жадная с UCB: Максимальный капитал = ${}, Среднее количество операций = {}'.format(max(results_UCB), mean(steps_UCB)))
```
